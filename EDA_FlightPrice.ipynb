```python
df['Day'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year']= df['Date_of_Journey'].str.split('/').str[2]
df.drop('Date_of_Journey',axis = 1,inplace = True)
```

```python
df.head()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Min'] = df['Dep_Time'].str.split(':').str[1]
df.drop('Dep_Time',axis = 1,inplace = True)
```

```python
df.head()
```

```python
df['Arr_Hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arr_Min']= df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop("Arrival_Time",axis = 1,inplace = True)
```

```python
df.head()
```

```python
df['D_Hour']= df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['D_Min'] = df['Duration'].str.split(' ').str[1].str.split('h').str[0]
df.drop('Duration',axis = 1,inplace = True)
```
